```python
import numpy as np
import h5py
import os

#The name of the clump catalog file
#    Cyclone is the name of the galaxy
#    RD0027 is the snapshot number it was run on
#     nref11c_nref9f refers to the resolution of the simulation
#        (11th level cooling refinement, 9th level forced refinement)
#    O_p1_density refers to the field that the clumps were identified on
#        (in this case O II, or singly ionized Oxygen, which traces cooler ionized material)
hierarchy_file = "./SASP_Tutorials/Cyclone_RD0027_nref11c_nref9fO_p1_density_ClumpTree.h5"

#Open the file
hf = h5py.File(hierarchy_file,'r')

#Each clump is a dataset in the hdf5 file labeled with a number
#You can get the leaf clump ids with:
leaf_ids = hf['leaf_clump_ids'][...]
print("Leaf ids are:",leaf_ids)
```

```python
#Let's look at the what info is in a clump object

lid = str(leaf_ids[0]) #just take the first clump
clump = hf[lid] #get the dataset
print("Here is a list of all pre-calculated keys:")
for key in clump.keys():
    print(key)
```

```python
###What does Each Key mean

####These are some technical keys related to the structure of the clump tree
#cell_ids - A list of the grid points in the simulation that make up the clump. This can be used to analyze the clump in the simulation itself
#shell_cell_ids - A list of cell ids for a 1 kpc shell of gas around the clump. Useful for characterizing environmental differences.
#center_disk_coords - The center of the clump in the frame of the galaxy
#parent_id - The clump id for the parent of this clump.
#child_ids - The clump ids for any children. Leaf clumps do not have any children, as they are the end of the hierarchy by definition
#tree_level - How far down the clump hierarchy is this


#### These keys some bulk properties of the clumps
#leaf_mass - The mass of the clump in units of Solar Masses
#leaf_temperature - The average temperature of the clump in units of K
#leaf_volume - The volume of the clump in units of kpc^3
#leav_vx - The average x-velocity of the clump in km/s. This is in the simulation frame, so you will probably want to convert it to the galactic frame 
#leaf_vy - The average y-velocity of the clump in km/s (simulation frame)
#leaf_vz - The average z-velocity of the clump in km/s (simulation frame)
#leaf_x - The x-coordinate of the clump center in units of kpc (simulation frame)
#leaf_y - The y-coordinate of the clump center in units of kpc (simulation frame)
#leaf_z - The z-coordinate of the clump in units of kpc (simulation frame)
#leaf_metallicity - The mass-weighted average metallicity of the clump
#leaf_pressure - The average pressure within the clump in Ba

#### These keys give the number densities of various species
#### All units are in cm^-3
#leaf_hi_num_dense - The average HI number density of the clump
#leaf_mgii_num_dense - The average Mg II number density
#leaf_oii_num_dense - The average O II number density
#leaf_oiii_num_dense - The average O III number density
#leaf_oiv_num_dense - The average O IV number density
#leaf_ov_num_dense - The average O V number density
#leaf_ovi_num_dense - The average O VI number density

#### These keys give some measures of emission, or how bright these clumps will be
### Emissivity is in units of photons / cm^3 / steradian
### Surface brightness are in units photons  / cm^2 / steradian
#leaf_cii_emissivity - The C II emissivity
#leaf_cii_sb_xlos - The CII Surface brightness, projected along the x-direction
#leaf_cii_sb_ylos - The CII Surface brightness, projected along the y-direction
#leaf_cii_sb_zlos - The CII Surface brightness, projected along the z-direction
#leaf_ciii_emissivity - The CIII emissivity
#leaf_ciii_sb_xlos - The CIII Surface brightness, projected along the x-direction
#leaf_ciii_sb_ylos - The CIII Surface brightness, projected along the y-direction
#leaf_ciii_sb_zlos - The CIII Surface brightness, projected along the z-direction
#leaf_civ_emissivity - The CIV emissivity
#leaf_civ_sb_xlos - The CIV Surface brightness, projected along the x-direction
#leaf_civ_sb_ylos - The CIV Surface brightness, projected along the x-direction
#leaf_civ_sb_zlos - The CIV Surface brightness, projected along the x-direction
#leaf_halpha_emissivity - The H-alpha emissivity
#leaf_halpha_sb_xlos - The H-alpha Surface brightness, projected along the x-direction
#leaf_halpha_sb_ylos - The H-alpha Surface brightness, projected along the x-direction
#leaf_halpha_sb_zlos - The H-alpha Surface brightness, projected along the x-direction
#leaf_lyalpha_emissivity - The Lyman-Alpha emissivity
#leaf_lyalpha_sb_xlos - The Lyman-Alpha Surface brightness, projected along the x-direction
#leaf_lyalpha_sb_ylos - The Lyman-Alpha Surface brightness, projected along the x-direction
#leaf_lyalpha_sb_zlos - The Lyman-Alpha Surface brightness, projected along the x-direction
#leaf_mgii_emissivity  - The Mg II emissivity
#leaf_mgii_sb_xlos - The Mg II Surface brightness, projected along the x-direction
#leaf_mgii_sb_ylos - The Mg II Surface brightness, projected along the y-direction
#leaf_mgii_sb_zlos - The Mg II Surface brightness, projected along the z-direction
#leaf_ovi_emissivity  - The O VI emissivity
#leaf_ovi_sb_xlos - The O VI Surface brightness, projected along the z-direction
#leaf_ovi_sb_ylos - The O VI Surface brightness, projected along the z-direction
#leaf_ovi_sb_zlos - The O VI Surface brightness, projected along the z-direction
#leaf_siii_emissivity  - The Si II emissivity
#leaf_siii_sb_xlos - The Si II Surface brightness, projected along the z-direction
#leaf_siii_sb_ylos - The Si II Surface brightness, projected along the z-direction
#leaf_siii_sb_zlos - The Si II Surface brightness, projected along the z-direction
#leaf_siiii_emissivity  - The Si III emissivity
#leaf_siiii_sb_xlos - The Si III Surface brightness, projected along the z-direction
#leaf_siiii_sb_ylos - The Si III Surface brightness, projected along the z-direction
#leaf_siiii_sb_zlos - The Si III Surface brightness, projected along the z-direction
#leaf_siiv_emissivity  - The Si IV emissivity
#leaf_siiv_sb_xlos - The Si IV Surface brightness, projected along the z-direction
#leaf_siiv_sb_ylos - The Si IV Surface brightness, projected along the z-direction
#leaf_siiv_sb_zlos - The Si IV Surface brightness, projected along the z-direction

####The following keys are the same as their corresponding clump keys, but for the shells around each clump
#shell_cooling_time - The cooling time of the material around the clump
#t_shear - The shearing time of the material around the clump
#shell_cii_emissivity
#shell_ciii_emissivity
#shell_civ_emissivity
#shell_halpha_emissivity
#shell_hi_num_dense
#shell_lyalpha_emissivity
#shell_mass
#shell_metallicity
#shell_mgii_emissivity
#shell_mgii_num_dense
#shell_oii_num_dense
#shell_oiii_num_dense
#shell_oiv_num_dense
#shell_ov_num_dense
#shell_ovi_emissivity
#shell_ovi_num_dense
#shell_pressure
#shell_siii_emissivity
#shell_siiii_emissivity
#shell_siiv_emissivity
#shell_temperature
#shell_volume
#shell_vx
#shell_vy
#shell_vz
```

```python
#Let's load a parameter and make a histogram
from matplotlib import pyplot as plt
leaf_masses = []
for lid in leaf_ids:
    #Loop through each leaf clump and append the mass to the array
    clump = hf[str(lid)]
    leaf_masses.append(clump['leaf_mass'][...])

plt.figure()
print("Smallest Clump=",np.min(leaf_masses),"Msun")
print("Largest Clump= ",np.max(leaf_masses),"Msun")

plt.hist(leaf_masses,bins=100,range=[200,200000])
plt.xlabel("Clump Mass [M$_{\\odot}$]",fontsize=16)
plt.ylabel("Number of Clumps",fontsize=16)
plt.show()
```
